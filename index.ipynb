```python
import random


def roll_dice():
    return random.randint(1, 6)


# Snakes and Ladders mapping
snakes = {99: 7, 65: 42, 25: 5, 87: 54, 91: 61, 47: 19}
ladders = {4: 56, 12: 50, 14: 55, 22: 58, 41: 79, 54: 88}


def move_player(position, dice_value):
    new_position = position + dice_value
    if new_position > 100:
        return position  # Stay in place if roll exceeds 100
    if new_position in snakes:
        print(
            f"Oops! Bitten by a snake at {new_position}. Down to {snakes[new_position]}."
        )
        return snakes[new_position]
    if new_position in ladders:
        print(
            f"Great! Climbed a ladder at {new_position}. Up to {ladders[new_position]}."
        )
        return ladders[new_position]
    return new_position


def play_game():
    player_positions = [1, 1]
    turn = 0  # 0 for Player 1, 1 for Player 2

    print("Welcome to Snakes and Ladders! üé≤")
    while True:
        input(f"Player {turn + 1}, press Enter to roll the dice...")
        dice_value = roll_dice()
        print(f"Player {turn + 1} rolled a {dice_value}")

        player_positions[turn] = move_player(player_positions[turn], dice_value)
        print(f"Player {turn + 1} is now at {player_positions[turn]}")

        if player_positions[turn] == 100:
            print(f"Congratulations! Player {turn + 1} wins! üèÜ")
            break

        turn = 1 - turn  # Switch turn


if __name__ == "__main__":
    play_game()
```

```python
import random


def roll_dice():
    return random.randint(1, 6)


snakes = {99: 7, 65: 42, 25: 5, 87: 54, 91: 61, 47: 19}
ladders = {4: 56, 12: 50, 14: 55, 22: 58, 41: 79, 54: 88}


def move(position, dice):
    new_pos = position + dice
    return (
        snakes.get(new_pos, ladders.get(new_pos, new_pos))
        if new_pos <= 100
        else position
    )


def play():
    pos = [1, 1]
    print("Welcome to Snakes and Ladders! üé≤")
    while True:
        for i in range(2):
            input(f"Player {i+1}, press Enter to roll...")
            dice = roll_dice()
            pos[i] = move(pos[i], dice)
            print(f"Player {i+1} rolled {dice}, now at {pos[i]}")
            if pos[i] == 100:
                print(f"Player {i+1} wins! üèÜ")
                return


if __name__ == "__main__":
    play()
```

```python
import random
import time


def roll_dice():
    return random.randint(1, 6)


snakes = {99: 7, 65: 42, 25: 5, 87: 54, 91: 61, 47: 19}
ladders = {4: 56, 12: 50, 14: 55, 22: 58, 41: 79, 54: 88}


def move(position, dice):
    new_position = position + dice
    if new_position > 100:
        print(f"Rolled {dice}, but stays at {position} (cannot exceed 100)")
        return position
    if new_position in snakes:
        print(f"Oh no! Snake at {new_position}, go down to {snakes[new_position]}")
        return snakes[new_position]
    if new_position in ladders:
        print(f"Yay! Ladder at {new_position}, climb to {ladders[new_position]}")
        return ladders[new_position]
    return new_position


def play():
    pos = [1, 1]
    print("üé≤ Welcome to Snakes and Ladders! üé≤")
    while True:
        i = 0
        while i < 2:
            extra_turn = True
            while extra_turn:
                input(f"Player {i+1}, press Enter to roll...")
                dice = roll_dice()
                print(f"Player {i+1} rolled a {dice}")
                new_pos = move(pos[i], dice)
                print(f"Player {i+1} moves from {pos[i]} to {new_pos}")
                pos[i] = new_pos
                if pos[i] == 100:
                    print(f"üéâ Player {i+1} wins! üéâ")
                    return
                extra_turn = dice == 6  # Player gets another turn if they roll a 6
                time.sleep(2)  # Pause for 5 seconds after each roll
            i += 1  # Move to next player


if __name__ == "__main__":
    play()
```
